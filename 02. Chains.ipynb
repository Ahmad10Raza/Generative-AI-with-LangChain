# chains In LangChain

 

**Chains in LangChain are sequences of components that work together to process user input, interact with large language models (LLMs), and generate responses.** They offer a modular way to structure and control the flow of information within LangChain applications.

**Key characteristics of chains:**

- **Linear structure:** Components are arranged sequentially, with each component performing a specific task and passing its output to the next component.
- **Flexibility:** Chains can be customized to create diverse application flows and adapt to different use cases.
- **Reusability:** Components can be reused in multiple chains, promoting code efficiency and modularity.

**Common components in chains:**

- **Prompts:** Define the starting point for LLM interactions, setting the context and instructions for the LLM.
- **Adapters:** Connect LangChain to specific LLMs, enabling communication and handling model-specific details.
- **Filters:** Process and modify text outputs, such as cleaning up formatting or removing unwanted content.
- **Memories:** Store and retrieve information across multiple interactions, providing context and continuity.

**Benefits of using chains:**

- **Structured development:** Promotes organized and maintainable code.
- **Flexibility and adaptability:** Facilitates experimentation with different workflows and LLMs.
- **Customization:** Allows tailoring applications to specific needs and use cases.


# Types Of Chains?

 **Here are some common types of chains in LangChain:**

**1. LLM Chain:**

- The most fundamental chain type.
- Takes an input, formats it, and passes it to an LLM for processing.
- Basic components: PromptTemplate, an LLM, and an optional output parser.

**2. Sequential Chain:**

- A series of chains executed in a specific order.
- Variations include:
    - SimpleSequentialChain: Singular input/output with the output of one step as input to the next.
    - SequentialChain: Allows for multiple inputs/outputs.

**3. Router Chain:**

- Routes input to different chains based on certain criteria.
- Useful for multi-functional applications or context-based responses.

**4. Index Chains:**

- Combine your data stored in indexes with LLMs.
- Examples include:
    - TextIndexChain: Uses a text index to retrieve relevant information for LLM responses.
    - DocumentStoreChain: Integrates with a document store for more complex data retrieval and interaction.

**5. BashChain:**

- A specialized chain for running filesystem commands.
- Allows interaction with external tools and data sources.

**6. Custom Chains:**

- You can create custom chains for specific functionalities and workflows.
- Extend LangChain's capabilities for tailored application development.


# 1. LLM Chains

## Deep Dive into LLM Chains

**Here's a comprehensive explanation of LLM Chains in LangChain, focusing on their structure, components, and functionalities:**

**Structure:**

- An LLM Chain is a **linear sequence of components** that collaborate to process user input, interact with an LLM, and generate responses.
- Each component performs a specific task and passes its output to the next in the chain.
- This modular approach allows for flexible and customizable workflows.

**Key Components:**

1. **PromptTemplate:** Defines the starting point for the chain, often using placeholders and user input to set the context and instructions for the LLM.
2. **LLM Adapter:** Bridges the gap between LangChain and the chosen LLM, handling model-specific details and communication.
3. **Optional Processors:** Depending on the workflow, you can add components like filters to clean up formatting or modify text outputs, memory to store and use context across interactions, or custom components for unique functionalities.
4. **Optional Output Processor:** Further formats and prepares the LLM's final response for presentation to the user.

**Functionalities:**

- **Basic Interactions:** LLM Chains enable basic communication with LLMs by taking user input, passing it to the model through the adapter, and presenting the response for user interaction.
- **Structured Workflows:** They facilitate building multi-step interactions by stringing together components. For example, an LLM Chain can:
    - Combine user input with data from an external source through an IndexChain.
    - Run specific processing steps or filter outputs before responding to the user.
    - Utilize context stored in a MemoryChain to generate personalized or consistent responses across multiple interactions.

**Benefits:**

- **Structured Development:** Promotes organized and maintainable code base due to the modular and sequential nature of chains.
- **Flexibility and Adaptability:** Allows for customizing and modifying chains to suit different use cases and LLM integrations.
- **Control and Optimization:** Enables fine-grained control over information flow within the application, optimizing output quality and efficiency.

**Example Use Case:**

Imagine a restaurant recommenation app powered by an LLM Chain. The chain might look like this:

1. **PromptTemplate:** "Tell me about a restaurant with *[cuisine]* food and a *[ambience]* atmosphere located in *[city]*. Be creative!"
2. **LLM Adapter:** Connects to an LLM like GPT-J and translates the prompt into model-specific format.
3. **MemoryChain:** Checks past user preferences and location data to personalize the LLM prompt further.
4. **LLM (GPT-J):** Generates a detailed description of a restaurant based on the prompt and context.
5. **Output Processor:** Formats the LLM response for readability and adds relevant links or images.

This is just one example, and LLM Chains can be adapted to handle various situations and tasks, making them a versatile tool for building LLM-driven applications.

# 2. Sequential Chain



**Sequential Chains enable you to execute multiple chains in a specific order, creating a multi-step workflow.** They facilitate more complex interactions and information processing than simple LLM Chains.

**Key Concepts:**

- **Chain of Chains:** A Sequential Chain acts as a container for other chains, orchestrating their execution in a predefined sequence.
- **Linear Flow:** The output of one chain becomes the input for the next, creating a structured flow of information.
- **Variations:**
    - **SimpleSequentialChain:** The most basic form, where each step has a single input and output.
    - **SequentialChain:** Offers more flexibility with multiple inputs and outputs at each step.

**Benefits of Using Sequential Chains:**

- **Structured Workflows:** Build complex interactions involving multiple steps, such as gathering data, processing it, and generating a response.
- **Modularity:** Break down complex tasks into smaller, more manageable chains, promoting code organization and reusability.
- **Flexibility:** Easily modify or add steps within the workflow as needed.
- **Experimentation:** Try different chains and configurations to optimize results.

**Common Use Cases:**

- **Summarizing a document:**
    1. IndexChain to retrieve relevant text sections.
    2. FilterChain to remove unnecessary content.
    3. LLMChain to generate a concise summary.
- **Generating creative text based on multiple prompts:**
    1. SequentialChain with multiple LLMChains, each using a different prompt.
    2. FilterChain to combine and refine the outputs.
- **Building interactive conversations:**
    1. MemoryChain to store context from previous interactions.
    2. LLMChain to generate responses based on user input and context.
    3. RouterChain to direct the conversation flow based on user choices.

**Example Python Code:**

```python
from langchain.chains import SimpleSequentialChain

chain1 = ...  # Define the first chain
chain2 = ...  # Define the second chain

sequential_chain = SimpleSequentialChain(chains=[chain1, chain2])
output = sequential_chain.run(input_data)
```

## Routers in LangChain: Navigating Your Application Flow

Router Chains in LangChain add a powerful layer of flexibility to your applications by enabling **conditional branching based on user input or other data**, directing the flow of information through alternative paths within your LangChain application. Here's a deep dive into their details:

**Concept:**

- A Router Chain acts as a decision point, evaluating a specified condition and directing the program execution based on the outcome.
- It takes an input, evaluates the condition, and routes the input to one of several **subchains** based on the result.

**Components:**

- **Condition:** Defined as a function that takes the input data and returns a boolean value.
- **Subchains:** Multiple chains connected to the Router, representing different execution paths depending on the condition.

**Benefits:**

- **Dynamic branching:** Tailor the application's response based on user input, context, or external data.
- **Complex workflows:** Handle diverse scenarios and situations within a single chain.
- **Reusability:** Break down complex logic into separate subchains for modularity and organization.
- **Improved user experience:** Adapt your application's behavior to user choices and preferences.

**Use Cases:**

- Building a chatbot that responds differently based on the user's sentiment.
- Creating a recommendation system that suggests diverse options based on user preferences.
- Developing an interactive story where user choices determine the narrative.

**Implementation:**

Here's a basic example of a Router Chain in Python:

```python
from langchain.chains import RouterChain

def is_positive_sentiment(text):
    # Implement sentiment analysis logic
    return True

chain1 = ...  # Chain for positive response
chain2 = ...  # Chain for negative response

router_chain = RouterChain(
    condition=is_positive_sentiment,
    chains={
        True: chain1,
        False: chain2
    }
)

text = "I loved the movie!"

response = router_chain.run(text)
# If is_positive_sentiment(text) returns True, response will be the output of chain1, otherwise response will be the output of chain2.
```

**Advanced Techniques:**

- **Nested Routers:** Combine multiple Router Chains for even more nuanced branching logic.
- **Custom Conditions:** Define complex conditions based on any relevant data or logic.
- **Default Chain:** Specify a fallback chain to handle situations not covered by other conditions.

Remember, Router Chains add significant power and flexibility to your LangChain applications. Experiment with different conditions and subchains to create dynamic and engaging user experiences.

````python
from langchain.chains.router import MultiPromptChain
from langchain.chains.router.llm_router import LLMRouterChain,RouterOutputParser
from langchain.prompts import PromptTemplate

llm = OpenAI(temperature=0.7, openai_api_key=OPENAI_API_KEY)
````

#### Defining Prompt Templates
Let’s consider a scenario where we need to direct inputs to specialized chains based on subjects such as Maths, Physics, History, or Computer Science. To accomplish this, we create distinct prompts for each subject: one for physics questions, another for math queries, a third for history inquiries, and a fourth for computer science-related matters. We meticulously design these prompts to cater to the unique needs of each subject area.

````python
physics_template = """You are a very smart physics professor. \
You are great at answering questions about physics in a concise\
and easy to understand manner. \
When you don't know the answer to a question you admit\
that you don't know.

Here is a question:
{input}"""


math_template = """You are a very good mathematician. \
You are great at answering math questions. \
You are so good because you are able to break down \
hard problems into their component parts,
answer the component parts, and then put them together\
to answer the broader question.

Here is a question:
{input}"""

history_template = """You are a very good historian. \
You have an excellent knowledge of and understanding of people,\
events and contexts from a range of historical periods. \
You have the ability to think, reflect, debate, discuss and \
evaluate the past. You have a respect for historical evidence\
and the ability to make use of it to support your explanations \
and judgements.

Here is a question:
{input}"""
````

````python
# Defining the prompt templates
prompt_infos = [
    {
        "name": "physics",
        "description": "Good for answering questions about physics",
        "prompt_template": physics_template
    },
    {
        "name": "math",
        "description": "Good for answering math questions",
        "prompt_template": math_template
    },
    {
        "name": "History",
        "description": "Good for answering history questions",
        "prompt_template": history_template
    }
]
````

#### Creating Destination Chains
Next, our focus shifts to crafting destination chains. These chains are activated by the RouterChain, functioning as individual language model chains, specifically LLM chains. Additionally, a default chain is outlined to handle situations where the router encounters ambiguity and cannot determine the suitable subchain to utilize. This default chain acts as a fallback option, ensuring a response even in cases of indecision.

````python
destination_chains = {}
for p_info in prompt_infos:
    name = p_info["name"]
    prompt_template = p_info["prompt_template"]
    prompt = ChatPromptTemplate.from_template(template=prompt_template)
    chain = LLMChain(llm=llm, prompt=prompt)
    destination_chains[name] = chain

destinations = [f"{p['name']}: {p['description']}" for p in prompt_infos]
destinations_str = "\n".join(destinations)
````

#### Creating a Multi-prompt Router Template


````python
MULTI_PROMPT_ROUTER_TEMPLATE = """Given a raw text input to a \
language model select the model prompt best suited for the input. \
You will be given the names of the available prompts and a \
description of what the prompt is best suited for. \
You may also revise the original input if you think that revising\
it will ultimately lead to a better response from the language model.

<< FORMATTING >>
Return a markdown code snippet with a JSON object formatted to look like:
```json
{{{{
    "destination": string \ name of the prompt to use or "DEFAULT"
    "next_inputs": string \ a potentially modified version of the original input
}}}}
```

REMEMBER: "destination" MUST be one of the candidate prompt \
names specified below OR it can be "DEFAULT" if the input is not\
well suited for any of the candidate prompts.
REMEMBER: "next_inputs" can just be the original input \
if you don't think any modifications are needed.

<< CANDIDATE PROMPTS >>
{destinations}

<< INPUT >>
{{input}}

<< OUTPUT (remember to include the ```json)>>"""
````

````python
default_prompt = ChatPromptTemplate.from_template("{input}")
default_chain = LLMChain(llm=llm, prompt=default_prompt)
````

### Creating Router Template

````python
router_template = MULTI_PROMPT_ROUTER_TEMPLATE.format(
    destinations=destinations_str
)
router_prompt = PromptTemplate(
    template=router_template,
    input_variables=["input"],
    output_parser=RouterOutputParser(),
)
````

````python
chain = MultiPromptChain(router_chain=router_chain,
                         destination_chains=destination_chains,
                         default_chain=default_chain, verbose=True
                        )
````

## BashChains: Powering Your LangChain App with Shell Commands

BashChains in LangChain open up a whole new avenue for your applications by letting you **leverage the power of the Linux shell**. Here's a breakdown of what they do and how they can enhance your projects:

**What are BashChains?**

- They're special components in LangChain that allow you to **execute system commands** like bash scripts or Linux CLI tools directly within your application workflow.
- This integration lets you **tap into external data sources, access file systems, and perform actions** that go beyond the typical capabilities of LLMs and traditional LangChain components.

**Unlocking Possibilities:**

- **Data Integration:** Access data stored in files, databases, or APIs accessible through shell commands.
- **External Tools:** Utilize powerful CLI tools like image processing utilities, text analysis algorithms, or scientific calculations.
- **Dynamic Workflows:** Combine the flow of information within LangChain with external actions for more versatile applications.
- **Automated Tasks:** Trigger automated actions like sending emails, generating reports, or managing file systems.

**Example Use Cases:**

- **Image-based recommendation system:** Use BashChain to run image recognition tools on uploaded photos and feed the results to an LLM for generating personalized recommendations.
- **Scientific data analysis:** Execute statistical analysis scripts on downloaded datasets and utilize the results to inform LLM-generated scientific summaries.
- **Automated report generation:** Combine data retrieved from different sources via BashChain commands and use LangChain to automatically generate reports based on the data.

**Important Considerations:**

- **Security:** Only use well-defined and trusted shell commands to avoid security vulnerabilities.
- **Complexity:** Keep BashChain logic simple and maintainable to avoid code spaghetti.
- **Error Handling:** Implement robust error handling for gracefully handling any issues with external commands.

**Getting Started:**

To use BashChain, simply import the `BashChain` class from LangChain and include it in your chain definition. Here's an example:

```python

# 6. Custom Chains

 **Unleashing Creativity with Custom Chains in LangChain**

Custom Chains in LangChain empower you to craft unique and tailored workflows, extending its capabilities beyond built-in chains. Here's how they work:

**Creating Tailored Workflows:**

- **Unbound by Constraints:** Define chains that perfectly suit your specific needs and application requirements, unconstrained by predefined structures.
- **Custom Components:** Build your own components using Python functions, classes, or external libraries, expanding LangChain's functionality.
- **Reusable Building Blocks:** Create reusable custom components for efficient development and structured code.

**Key Steps to Create a Custom Chain:**

1. **Define Your Component:**
   - Write a Python function or class that encapsulates the desired functionality.
   - Ensure it takes input data, processes it, and returns output in a consistent format.
2. **Instantiate Your Chain:**
   - Create a `Chain` object from LangChain.
   - Include your custom component as one of its components.
3. **Run the Chain:**
   - Pass input data to the chain's `run()` method.
   - The chain will execute its components in order, passing data between them.

**Example:**

```python
import langchain

def my_custom_component(input_data):
    # Perform custom processing on the input data
    processed_data = input_data.upper()  # Simple example
    return processed_data

chain = langchain.Chain(
    components=[
        my_custom_component,
        # Other LangChain components...
    ]
)

output = chain.run("hello world")  # Output: "HELLO WORLD"
```

**Unlocking Possibilities:**

**Benefits:**

- **Flexibility:** Adapt LangChain to any project or idea.
- **Reusability:** Build a library of custom components for efficient development.
