# Alpaca LLM?

Alpaca LLM (Large Language Model) refers to a **fine-tuned version of the LLaMA 7B model**, also known as "Jurassic-1 Jumbo" or "Jurassic-1 Fossil". It stands out for its focus on **improved factual accuracy and safety** compared to the original LLaMA, while retaining its strong language generation capabilities.

Here's a deeper dive into Alpaca LLM:

**Key Features:**

* **Accuracy and Factuality:** Trained on a dataset emphasizing factual accuracy, reducing the risk of misinformation and unreliable outputs.
* **Safety and Ethics:** Employs various techniques to mitigate bias and harmful content, promoting responsible AI usage.
* **Open-Source Availability:** Freely accessible to researchers and developers, fostering transparency and collaboration.
* **Strong Language Generation:** Continues to excel in tasks like generating creative text formats, writing different kinds of creative content, and translating languages.

**Benefits and Applications:**

* **Enhanced Reliability:** Ideal for tasks requiring accurate and dependable information, such as research, document summarization, and data analysis.
* **Ethical and Responsible AI:** Promotes safe and responsible AI development, reducing the risk of harmful or biased outputs.
* **Accessibility and Research:** Open-source availability benefits researchers and developers, supporting further advancements in NLP and LLM technology.
* **Versatile Applications:** Applicable to various creative and informative tasks, making it a valuable tool for diverse fields.

**Comparison to LLaMA:**

* Alpaca retains LLaMA's strong language generation capabilities.
* Offers significant improvements in factual accuracy and safety compared to LLaMA.
* Open-source approach stands in contrast to LLaMA's initial restricted availability.

**Limitations and Considerations:**

* Still under development, and further improvements in accuracy and safety might be needed.
* Requires careful handling and ethical considerations to avoid potential misuse.
* Depending on the specific use case, other LLMs might offer better performance for tasks not prioritized by Alpaca.

**Overall, Alpaca LLM emerges as a promising advancement in LLM technology, addressing concerns about factual accuracy and safety while maintaining strong language generation capabilities. Its open-source nature further contributes to responsible AI development and broadens its accessibility for research and diverse applications.**



# LLaMA?

LLaMA, which stands for **Large Language Model Meta AI**, is a state-of-the-art foundational large language model developed by Meta AI. Here's a breakdown of its key features and capabilities:

**1. Generative Power:**

- LLaMA boasts impressive language generation capabilities, excelling at tasks like:
    - Writing different creative text formats like poems, code, scripts, musical pieces, emails, letters, etc.
    - Translating languages with high accuracy and fluency.
    - Answering your questions in an informative way, even if they are open ended, challenging, or strange.

**2. Open-Source Focus:**

- LLaMA is a major step towards open-sourcing large language models, promoting transparency and accessibility in the field of AI.
- This allows researchers and developers to work directly with the model, leading to further advancements and broader applications.

**3. Size and Parameters:**

- LLaMA comes in two versions:
    - LLaMA 137B with 137 billion parameters, offering a balance between performance and resource usage.
    - LLaMA 65B with 65 billion parameters, requiring less hardware resources but still exhibiting strong capabilities.

**4. Ethical Considerations:**

- While LLaMA demonstrates powerful language abilities, it's important to be aware of potential biases and safety concerns inherent in LLMs.
- Using LLaMA responsibly and implementing careful checks for harmful content is crucial to ensure ethical AI development and usage.

**5. Applications:**

- LLaMA's versatility opens doors to various applications in diverse fields, including:
    - Creative writing and content generation.
    - Education and personalized learning.
    - Research and data analysis.
    - Machine translation and communication.

**Limitations and Ongoing Research:**

- Like other LLMs, LLaMA is still under development, and further improvements are required in areas like factual accuracy and reasoning.
- Researchers are actively working on mitigating biases and safety issues, aiming to make LLMs more reliable and ethical tools.

**Overall, LLaMA represents a significant milestone in open-source LLM development, offering powerful language processing capabilities and promoting transparency in the field. Its potential applications are vast, while continued research and ethical considerations are necessary to ensure responsible and beneficial use of this advanced technology.**

## Loading Alpaca7B

```python
from transformers import LlamaTokenizer, LlamaForCausalLM, GenerationConfig, pipeline
from langchain.llms import HuggingFacePipeline
from langchain import PromptTemplate, LLMChain
from langchain import PromptTemplate, LLMChain

import torch
```

```python
tokenizer = LlamaTokenizer.from_pretrained("chavinlo/alpaca-native")

base_model = LlamaForCausalLM.from_pretrained(
    "chavinlo/alpaca-native",
    load_in_8bit=True,
    device_map='auto',
)
```

```python
pipe = pipeline(
    "text-generation",
    model=base_model,
    tokenizer=tokenizer,
    max_length=256,
    temperature=0.6,
    top_p=0.95,
    repetition_penalty=1.2
)

local_llm = HuggingFacePipeline(pipeline=pipe)
```

```python
template = """respond to the instruction below. behave like a chatbot and respond to the user. try to be helpful.
### Instruction:
{instruction}
Answer:"""
prompt = PromptTemplate(template=template, input_variables=["instruction"])
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
llm_chain = LLMChain(prompt=prompt,
                     llm=local_llm
                     )
question = "Tell me about India?"
print(llm_chain.run(question))
```

```python
question2="Who  is Prime Minister of India?"
print(llm_chain.run(question2))
```

## Setting Up Chatbot with Memory:

```python
from langchain.chains import ConversationChain
from langchain.chains.conversation.memory import ConversationBufferWindowMemory
```

```python
mem=ConversationBufferWindowMemory(k=7)

conversation=ConversationChain(llm=local_llm,
                               verbose=True,
                               memory=mem)
```

```python
conversation.prompt.template
```

```python
conversation.prompt.template = '''The following is a friendly conversation between a human and an AI called Alpaca. The AI is talkative and provides lots of specific details from its context. If the AI does not know the answer to a question, it truthfully says it does not know.

Current conversation:
{history}
Human: {input}
AI:'''
```

```python
conversation.predict(input="Hi there! I am Nikola Tesla")
```

```python
conversation.predict(input="Can you tell me what an Alpaca is?")
```

```python
conversation.predict(input="How is it different from a Llama?")
```

```python
conversation.predict(input="Can you give me some good names for a pet llama?")
```
